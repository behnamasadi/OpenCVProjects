# Perspective projection

Perspective projection (also called perspective transformation) projects three-dimensional objects onto a picture plane. This has the effect that distant objects appear smaller than nearer ones, and lines that are parallel in the scene (meeting only at a point at infinity) appear to converge in the projected image.

A perspective transformation corresponds to a change of viewpoint. It preserves **collinearity** and **incidence** (straight lines stay straight) and the **cross-ratio** of four collinear points, but **parallelism, length, and angles are not preserved**. Although the mapping is linear in *homogeneous* coordinates (a single 3×3 matrix), it is **nonlinear** in ordinary Cartesian image coordinates because of the perspective division (dividing by the third coordinate).



depending on the orientation of the projection plane towards the axes of the depicted object, perspective projection is categorized into **one-point**, **two-point** and **three-point** perspective.

## One Point Perspective
In one point perspective all lines recede toward one vanishing point.

Perspective Transformation Matrix has the following form:

<img src="https://latex.codecogs.com/svg.image?\displaystyle&space;{\begin{bmatrix}M_{11}&space;&M_{12}&space;&M_{13}&space;\\M_{21}&space;&M_{22}&space;&M_{23}&space;\\M_{31}&space;&M_{32}&1\end{bmatrix}}" title="https://latex.codecogs.com/svg.image?\displaystyle {\begin{bmatrix}M_{11} &M_{12} &M_{13} \\M_{21} &M_{22} &M_{23} \\M_{31} &M_{32}&1\end{bmatrix}}" />  

This 3×3 matrix is a **homography** (a projective transform). It has **8 degrees of freedom**: it holds 9 entries, but is defined only up to scale, so one entry is normalized (here $M_{33}=1$). A perspective transform maps any **quadrilateral onto any other quadrilateral**, which is why 4 point correspondences (4 pairs → 8 equations) are enough to determine it uniquely.

> **Note:** this 3×3 matrix is the same object described in the companion notebook [`homography.ipynb`](homography.ipynb), which covers estimating it from many point pairs (`cv::findHomography`), RANSAC, and planar-scene geometry. This notebook focuses on the projective transform itself and OpenCV's image-warping functions.

# OpenCV Perspective Functions

## Estimating the matrix: `cv::getPerspectiveTransform`
Given **4 pairs** of corresponding points (source and destination), `cv::getPerspectiveTransform` computes the 3×3 perspective matrix that maps the source quadrilateral onto the destination quadrilateral. When you have more than 4 correspondences and want a least-squares or robust (RANSAC) fit, use `cv::findHomography` instead — see [`homography.ipynb`](homography.ipynb).

## Applying the matrix: `cv::warpPerspective`
The function `cv::warpPerspective` transforms the source image using the specified matrix:

## Typical Uses
- **Bird's-eye (top-down) view:** re-project a slanted ground plane (a road, a table) so it appears as if viewed from directly above.
- **Document scanning / dewarping:** map the four corners of a photographed document onto a clean rectangle to obtain a flat, front-on scan.
- **Rectifying planar surfaces** (signs, whiteboards, book pages) before OCR or measurement.
