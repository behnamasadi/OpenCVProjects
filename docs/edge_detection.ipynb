## Scharr

For a 3×3 aperture the Sobel kernel is only a coarse approximation of the derivative-of-Gaussian. The **Scharr** operator has better rotational symmetry (smaller angular error), so it yields more accurate gradient directions at the same 3×3 size:

```
G_x = | +3   0   -3 |        G_y = | +3  +10   +3 |
      | +10  0  -10 |              |  0    0    0 |
      | +3   0   -3 |              | -3  -10   -3 |
```

## Computing gradients in OpenCV

```cpp
// Use a signed ddepth (CV_16S / CV_32F / CV_64F) so negative gradients are not clipped.
void cv::Sobel(InputArray src, OutputArray dst, int ddepth,
               int dx, int dy, int ksize = 3,
               double scale = 1, double delta = 0,
               int borderType = BORDER_DEFAULT);
```

- `dx`, `dy` are the derivative orders, e.g. `dx=1, dy=0` for the horizontal derivative.
- `ksize` is the aperture size (1, 3, 5, 7); passing `ksize = cv::FILTER_SCHARR` (-1) selects the Scharr kernel, which is also exposed directly as `cv::Scharr(src, dst, ddepth, dx, dy, ...)`.
- Compute `Gx` and `Gy` separately, then combine the magnitude with `cv::magnitude` (exact L2), or `cv::addWeighted(abs(Gx), 0.5, abs(Gy), 0.5, 0, dst)` for a fast L1 approximation, and use `cv::convertScaleAbs` before display.

The Laplacian sums the two second derivatives, `∇²L = L_xx + L_yy`, which gives the common 3×3 aperture (4-connected). An 8-connected variant that also weights the diagonals is often used:

# Difference of Gaussians (DoG) 
Convolution Properties: Associativity

<!-- 
<img src="https://latex.codecogs.com/svg.latex?{\displaystyle f*(g*h)=(f*g)*h}" /> 
-->

<img src="https://latex.codecogs.com/svg.image?\inline&space;{\displaystyle&space;f*(g*h)=(f*g)*h}" title="https://latex.codecogs.com/svg.image?\inline {\displaystyle f*(g*h)=(f*g)*h}" />

Because differentiation and Gaussian smoothing are both linear, their order can be swapped: instead of smoothing an image and then differentiating it, we can convolve the image once with a pre-computed **derivative of Gaussian** kernel, which is cheaper and less noise-sensitive.

The **Difference of Gaussians** subtracts two Gaussian-blurred versions of an image at different scales (σ₁ < σ₂). The result is a band-pass filter that closely approximates the Laplacian of Gaussian (below), which is why DoG is widely used as a fast substitute for LoG.

# Laplacian of Gaussian (LoG)

Applying the noise-sensitive Laplacian directly is impractical, so the image is first smoothed with a Gaussian and then differentiated. Because both operations are linear convolutions they can be merged into a single **Laplacian-of-Gaussian** kernel:

<img src="https://latex.codecogs.com/svg.image?\inline&space;\text{LoG}(x,y)=-\frac{1}{\pi\sigma^4}\left[1-\frac{x^2&plus;y^2}{2\sigma^2}\right]e^{-\frac{x^2&plus;y^2}{2\sigma^2}}" title="LoG" />

This is the **Marr–Hildreth** operator; its rotationally symmetric "Mexican-hat" shape means it responds to intensity changes in every direction. Edges are located at the **zero crossings** of the LoG response (points where it changes sign between neighbouring pixels), and σ controls the scale of the detected edges. Weak crossings can be rejected by requiring a minimum gradient (slope) across the crossing. As noted above, DoG is a common fast approximation of LoG.

# Canny Edge

The Canny detector is a multi-stage pipeline designed to be optimal for step edges (good detection, good localization, single response per edge):

1. **Gaussian smoothing** — blur the image to suppress noise; the kernel size / σ trades noise rejection against edge localization.
2. **Gradient magnitude and direction** — compute `Gx`, `Gy` (typically with Sobel), then `|G| = sqrt(Gx² + Gy²)` and `θ = atan2(Gy, Gx)`.
3. **Non-maximum suppression** — thin the ridges by keeping a pixel only if its magnitude is a local maximum along the gradient direction (quantized to 0°, 45°, 90°, 135°). This produces 1-pixel-wide edges.
4. **Double threshold** — classify pixels using an upper and a lower threshold: magnitude ≥ high → **strong** edge, between low and high → **weak** edge, < low → suppressed.
5. **Edge tracking by hysteresis** — keep every strong edge, and keep a weak edge only if it is connected (8-connectivity) to a strong edge; discard the rest. This links genuine edges while dropping isolated noise responses.

```cpp
void cv::Canny(InputArray image, OutputArray edges,
               double threshold1, double threshold2,
               int apertureSize = 3, bool L2gradient = false);
```

- `threshold1` / `threshold2` are the lower and upper hysteresis thresholds (order does not matter, the smaller is treated as low). Canny recommended a high:low ratio of about 2:1 to 3:1.
- `apertureSize` is the Sobel aperture used for the gradient (3, 5 or 7).
- `L2gradient`: if `true` the magnitude uses the exact L2 norm `sqrt(Gx² + Gy²)`; if `false` (default) it uses the cheaper L1 norm `|Gx| + |Gy|`.
