Since the relative pose of gripper and camera is fixed, the constant transform `X = T_c^g` is the same for every pose pair `(i, j)`, which is exactly what we solve for.

To obtain a well-conditioned solution you need several robot poses (at least 3, in practice 10 or more) whose relative motions have **sufficiently varied rotation axes** — pure translations, or rotations about a single fixed axis, leave `X` under-determined.

`cv::calibrateHandEye` provides five solvers, chosen via the `method` argument: `CALIB_HAND_EYE_TSAI` (default), `CALIB_HAND_EYE_PARK`, `CALIB_HAND_EYE_HORAUD`, `CALIB_HAND_EYE_ANDREFF`, and `CALIB_HAND_EYE_DANIILIDIS`. Its inputs are the per-pose gripper-to-base rotation/translation (`R_gripper2base`, `t_gripper2base`, from robot forward kinematics) and the target-to-camera rotation/translation (`R_target2cam`, `t_target2cam`, from camera extrinsic calibration); it outputs `R_cam2gripper`, `t_cam2gripper`, i.e. the camera-to-gripper transform `X = {}^{g}T_{c}`.



Refs: [1](https://docs.opencv.org/4.x/d9/d0c/group__calib3d.html#gaebfc1c9f7434196a374c382abf43439b), [2](https://www.youtube.com/watch?v=xQ79ysnrzUk), [3](https://campar.in.tum.de/Chair/HandEyeCalibration)

## 1.2 Eye-to-Hand

In the **eye-to-hand** configuration the camera is fixed in the workspace (mounted on the robot base or a static stand) and the calibration target is attached to the gripper — the opposite of eye-in-hand, where the camera moves with the gripper. The unknown `X` is then the camera-to-base transform `{}^{b}T_{c}`. `cv::calibrateHandEye` handles this case too: pass the inverted robot poses (base-to-gripper, `{}^{g}T_{b}`) in place of `R_gripper2base`/`t_gripper2base`, and the returned transform is the camera pose expressed in the base frame.

## 1.3 OpenCV Eye-in-Hand API

The following API computes Hand-Eye calibration:  <img src="https://latex.codecogs.com/svg.latex?_%7B%7D%5E%7Bg%7D%5Ctextrm%7BT%7D_c" alt="_{}^{g}\textrm{T}_c" />







```cpp
calibrateHandEye	(	InputArrayOfArrays 	R_gripper2base,
InputArrayOfArrays 	t_gripper2base,
InputArrayOfArrays 	R_target2cam,
InputArrayOfArrays 	t_target2cam,
OutputArray 	R_cam2gripper,
OutputArray 	t_cam2gripper,
HandEyeCalibrationMethod 	method = CALIB_HAND_EYE_TSAI 
)	
```
