# Affine Transformation
An affine transformation is a geometric transformation that preserves **collinearity** (points on a line stay on a line), **parallelism** (parallel lines stay parallel), and **ratios of lengths along a line** (the midpoint of a segment stays the midpoint), but does **not** in general preserve distances (lengths) or angles.

Examples of affine transformations include translation, scaling, similarity, reflection, rotation, shear mapping, and any composition of them.

# Affine Map
An affine map is the composition of two functions: a **linear map** followed by a **translation**. If the linear map is represented by an (invertible) matrix <img src="https://latex.codecogs.com/svg.image?A" title="https://latex.codecogs.com/svg.image?A" /> and the translation by the addition of a vector <img src="https://latex.codecogs.com/svg.image?b" title="https://latex.codecogs.com/svg.image?b" />, then an affine map <img src="https://latex.codecogs.com/svg.image?f" title="https://latex.codecogs.com/svg.image?f" /> acting on a vector <img src="https://latex.codecogs.com/svg.image?x" title="https://latex.codecogs.com/svg.image?x" /> can be written as

<img src="https://latex.codecogs.com/svg.image?{\displaystyle&space;\mathbf&space;{y}&space;=f(\mathbf&space;{x}&space;)=A\mathbf&space;{x}&space;&plus;\mathbf&space;{b}&space;.}" title="https://latex.codecogs.com/svg.image?{\displaystyle \mathbf {y} =f(\mathbf {x} )=A\mathbf {x} +\mathbf {b} .}" />

# Affine Transformation Matrix
Using an **augmented (homogeneous) matrix**, both the linear part and the translation can be expressed as a single matrix multiplication:

<img src="https://latex.codecogs.com/svg.image?{\displaystyle&space;{\begin{bmatrix}\mathbf&space;{y}&space;\\1\end{bmatrix}}=\left[{\begin{array}{ccc|c}&A&&\mathbf&space;{b}&space;\\0&\cdots&space;&0&1\end{array}}\right]{\begin{bmatrix}\mathbf&space;{x}&space;\\1\end{bmatrix}}}" title="https://latex.codecogs.com/svg.image?{\displaystyle {\begin{bmatrix}\mathbf {y} \\1\end{bmatrix}}=\left[{\begin{array}{ccc|c}&A&&\mathbf {b} \\0&\cdots &0&1\end{array}}\right]{\begin{bmatrix}\mathbf {x} \\1\end{bmatrix}}}" />

In 2D, the top two rows form a **2&times;3 matrix** <img src="https://latex.codecogs.com/svg.image?[A\,|\,b]" title="[A|b]" /> with **6 degrees of freedom** (the four entries of the 2&times;2 linear part <img src="https://latex.codecogs.com/svg.image?A" title="A" /> plus the two translation entries in <img src="https://latex.codecogs.com/svg.image?b" title="b" />). The appended bottom row is always fixed at <img src="https://latex.codecogs.com/svg.image?[0\;0\;1]" title="[0 0 1]" />, which is exactly what keeps the map affine (this is the row that a full perspective transform is allowed to change; see below).

## Affine vs. Perspective (Homography)
A **perspective transformation (homography)** is described by a full 3&times;3 matrix whose bottom row is a general <img src="https://latex.codecogs.com/svg.image?[h_{31}\;h_{32}\;h_{33}]" title="[h31 h32 h33]" /> (8 degrees of freedom, up to scale), and the result must be normalized by dividing by the third homogeneous coordinate. The key differences:

- **Affine**: bottom row fixed at <img src="https://latex.codecogs.com/svg.image?[0\;0\;1]" title="[0 0 1]" />, **6 DOF**, parallel lines stay parallel.
- **Perspective**: bottom row free, **8 DOF**, parallel lines may converge to a vanishing point (only straight lines are preserved, not parallelism).

Every affine transform is a special case of a homography, but not vice versa.

## OpenCV Affine Transformation
In OpenCV you can call `cv::warpAffine` to apply an affine transformation to an image. The function transforms the source image using the specified 2&times;3 matrix <img src="https://latex.codecogs.com/svg.image?M" title="M" />:

<img src="https://latex.codecogs.com/svg.image?dst&space;(x,y)&space;=&space;src&space;(&space;M&space;_{11}&space;x&space;&plus;&space;M&space;_{12}&space;y&space;&plus;&space;M&space;_{13},&space;M&space;_{21}&space;x&space;&plus;&space;M&space;_{22}&space;y&space;&plus;&space;M&space;_{23})" title="https://latex.codecogs.com/svg.image?dst (x,y) = src ( M_{11} x + M_{12} y + M_{13}, M_{21} x + M_{22} y + M_{23})" />

The output pixel coordinates are an affine (linear-plus-translation) function of the input coordinates. When you apply an affine transform to an image the output size may differ from the input, so you have to specify it:

`cv::warpAffine( src, dst, affine_mat, cv::Size(width, height) );`

For pure scaling you can also use `cv::resize`, and to build a rotation-about-a-point matrix use `cv::getRotationMatrix2D`.

# Finding the Affine Transform
- `cv::getAffineTransform` computes the exact 2&times;3 affine matrix from **3 point correspondences** (3 pairs of source/destination points uniquely determine the 6 parameters).
- `cv::estimateAffine2D` robustly estimates a full 6-DOF affine transform from many (possibly noisy) correspondences, using RANSAC/LMEDS to reject outliers.
- `cv::estimateAffinePartial2D` estimates a restricted 4-DOF affine transform (rotation + uniform scale + translation, i.e. a similarity transform, with no shear or non-uniform scaling).

(Note: `cv::estimateAffine3D` is the 3D counterpart — it estimates a 3D affine transform between two point sets, not a 2D image transform.)
